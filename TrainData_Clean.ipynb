```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_excel('/content/drive/MyDrive/Data/05 Input Data/Prediction of Placement Status Data/01 Train Data.xlsx')
train.head()
```

```python
train.drop(train.columns[[2,3,5,6,7,8,9,10,11,12,13,15,16,18]],axis = 1,inplace = True)
```

```python
train.isna().sum()
```

```python
train.dropna(subset=['College Name','Placement Status'],inplace = True)
train.head()
```

```python
train.dtypes
```

```python
categorical_values=['First Name','Email ID','Ticket Type','College Name','Designation','Placement Status']
```

```python
train[categorical_values].nunique()
```

```python
train_clean = train.drop(['First Name','Email ID','College Name','Designation','Ticket Type'],axis=1)
```

```python
train_clean['Placement Status'] = train_clean['Placement Status'].map({'Placed':1,'Not placed':0})
train_clean.head()
```

```python
train_clean = pd.get_dummies(train_clean)
```
